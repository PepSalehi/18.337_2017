```julia
function rank_approx(M, k)
    U, S, V = svd(M)   
    M = U[:, 1:k] * Diagonal(S[1:k]) * V[:, 1:k]'   
    M = min.(max.(M, 0.0), 1.)
end

n = 100
#@manipulate for k1 in 1:n, k2 in 1:n, k3 in 1:n
@manipulate for k in 1:n
    colorview( # RGB, 
        Gray,
                rank_approx(channels[:,:], k) #,
               # rank_approx(channels[2,:,:], k),
                #rank_approx(channels[3,:,:], k)
    )
end
```
